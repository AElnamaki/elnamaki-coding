## connectivity and structure 

## N = 8 

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the sequences representing paths in the graph
sequences = [
    [0, 1, 1, 2, 3, 5, 8],
    [0, 2, 2, 4, 6],
    [0, 3, 3, 6],
    [0, 4, 4, 8],
    [0, 5, 5],
    [0, 6, 6],
    [0, 7, 7],
    [0, 8, 8],
    [1, 3, 4, 7],
    [1, 4, 5],
    [1, 5, 6],
    [1, 6, 7],
    [1, 7, 8],
]

# Create a directed graph
G = nx.DiGraph()
```

```python
# Add edges from each sequence
for seq in sequences:
    for i in range(len(seq) - 1):
        G.add_edge(seq[i], seq[i + 1])

# Analysis
# Number of nodes and edges
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

# In-degree and out-degree for each node
in_degrees = dict(G.in_degree())
out_degrees = dict(G.out_degree())

# Strongly connected components
scc = list(nx.strongly_connected_components(G))
num_scc = len(scc)
```

```python
# Print analysis
print(f"Number of nodes: {num_nodes}")
print(f"Nodes: {sorted(G.nodes())}")
print(f"Number of edges: {num_edges}")
print("In-degrees:", in_degrees)
print("Out-degrees:", out_degrees)
print(f"Number of strongly connected components: {num_scc}")
print("Strongly connected components:", scc)

# Visualize the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)  # Consistent layout
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, arrowsize=15)
plt.title("Graph of Nodes Up to 8 from Maximal Seed Sequences")
plt.show()
```
