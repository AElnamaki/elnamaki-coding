## â™»ï¸ Symmetric Fibonacci Sequence (Negafibonacci)

In classical arithmetic, the **Fibonacci sequence** is defined only for non-negative indices:

$$
F_0 = 0, \quad F_1 = 1, \quad F_n = F_{n-1} + F_{n-2} \text{ for } n \geq 2
$$

However, the Fibonacci recurrence relation can be extended **symmetrically** into the negative index domain.  
This generates the **Negafibonacci sequence**, satisfying:

$$
F_{-n} = (-1)^{n+1} \cdot F_n
$$

This anti-symmetric recursion creates a mirrored sequence:

- **Positive Fibonacci**: Grows forward via addition.
- **Negative Fibonacci**: Mirrors backward with sign alternation.

### ðŸ” Symmetric Fibonacci in Code

We define a function to compute the full **symmetric Fibonacci span** from $-n$ to $+n$:


```python
def symmetric_fibonacci_sequence(depth: int) -> list:
    # Compute positive Fibonacci up to depth
    pos = [0, 7]
    for _ in range(2, depth + 1):
        pos.append(pos[-1] + pos[-2])

    # Compute negative (negafibonacci)
    neg = []
    for i in range(1, depth + 1):
        val = (-1) ** (i + 1) * pos[i]
        neg.insert(0, val)  # prepend to the list

    return neg + pos

# Example usage
depth = 5
sequence = symmetric_fibonacci_sequence(depth)
indices = list(range(-depth, depth + 1))
depths = [abs(n) for n in indices]  # Use absolute index as depth layer
print(sequence)
print(indices)
print(depths)
```

```python
import matplotlib.pyplot as plt

# === 2D Plot with Annotations ===
plt.figure(figsize=(10, 5))
plt.plot(indices, sequence, marker='o', color='blue')

# Annotate each point with its value
for i, (x, y) in enumerate(zip(indices, sequence)):
    plt.text(x, y + 5, str(y), ha='center', fontsize=9, color='darkred')  # Offset +5 for clarity

plt.title(f"2D Symmetric Fibonacci Sequence (Depth={depth})")
plt.xlabel("Index (n)")
plt.ylabel("F(n)")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
# === 3D Plot ===
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(indices, sequence, depths, marker='o', color='purple', linewidth=2)

ax.set_title(f"3D Symmetric Fibonacci (Seed=0,1 | Depth={depth})")
ax.set_xlabel("Index (n)")
ax.set_ylabel("F(n)")
ax.set_zlabel("Depth |n|")
plt.tight_layout()
plt.show()
```
